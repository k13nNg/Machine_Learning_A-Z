# Import the modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the dataset


```python
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning A - Z/Part 3 - Classification/Classification Summative Project/drug200.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

# Encoding Categorical Data


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

ct_X = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1, 2, 3])], remainder='passthrough')
X = np.array(ct_X.fit_transform(X))

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# np.set_printoptions(threshold=np.inf)
print(X)
# print(X[:, -2:])
```

```python
print(y)
```

# Splitting the dataset into Training and Testing sets


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
print(X_train)
```

# Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, -2:] = sc.fit_transform(X_train[:, -2:])
X_test[:, -2:] = sc.transform(X_test[:, -2:])
```

```python

print(X_train)
```

# Fitting the model to the Training set

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(max_iter = 10000)
classifier.fit(X_train, y_train)
```

# Predicting the Test Set

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

# Evaluating the model

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
