# Import the modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning A - Z/Part 2 - Regression/Regression Summative Project/boston.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
y = y.reshape(len(y),1)
```

# Splitting the dataset into the Training and Testing dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

# Training the Support Vector Regression model

```python
from sklearn.svm import SVR

regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

# Model Predictiona and Evaluation

```python
from sklearn.metrics import r2_score

y_pred = regressor.predict(X_test)

r2_score(y_test, y_pred)
```
